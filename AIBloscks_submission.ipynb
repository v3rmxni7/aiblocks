```python
import pandas as pd
import json
import re
import ast
from tqdm import tqdm

print("Before column selection:", df.shape)
print("Columns:", df.columns.tolist())

df = df[["ResumeText", "GPT_Output"]].copy()
df.columns = ["resume_text", "gpt_output"]

print("After column selection:", df.shape)
df.head(3)
```

```python
# 2. BASIC RESUME CLEANING

def clean_resume(text: str) -> str:
    if not isinstance(text, str):
        return ""


    text = text.replace("{new_line}", "\n")


    text = re.sub(r"\n\s*\n+", "\n\n", text)


    text = re.sub(r"[ ]{2,}", " ", text)

    return text.strip()


df["resume_text"] = df["resume_text"].apply(clean_resume)
print("Cleaned resume_text. Example:")
df["resume_text"].iloc[0][:500]
```

```python
# 3A. SAFE PARSING OF GPT OUTPUT

def safe_load_json(raw):
    """
    GPT_Output can be:
    - a real dict
    - a string repr of dict (single quotes)
    - JSON string
    - NaN / None
    """
    if isinstance(raw, dict):
        return raw

    if raw is None:
        return None

    # handle NaN
    if isinstance(raw, float) and pd.isna(raw):
        return None

    text = str(raw).strip()
    if not text:
        return None

    try:
        return ast.literal_eval(text)
    except Exception:
        pass

    try:
        return json.loads(text)
    except Exception:
        return None
```

```python
# 3B. DATE NORMALIZATION

from dateutil import parser

MONTH_RE = re.compile(r"(jan|feb|mar|apr|may|jun|jul|aug|sep|oct|nov|dec)", re.IGNORECASE)

def normalize_date(d):
    """
    Normalize only when safe.
    Target format: YYYY-MM
    Special case: 'Present' / 'Till date' → 'Present'
    If we are not confident, return original string.
    """
    if d is None:
        return ""

    s = str(d).strip()
    if not s:
        return ""

    low = s.lower()

    if "present" in low or "till date" in low or "till-date" in low or "currently" in low:
        return "Present"

    if re.fullmatch(r"\d{4}", s):
        return s

    if re.search(r"\d{4}\s*[-–]\s*\d{4}", s):
        return s

    has_month_name = bool(MONTH_RE.search(low))
    has_mm_yyyy = bool(re.fullmatch(r"\d{1,2}/\d{4}", s))

    if not (has_month_name or has_mm_yyyy):
        return s

    try:
        dt = parser.parse(s, fuzzy=True, dayfirst=False)
        return dt.strftime("%Y-%m")
    except Exception:
        return s
```

```python
# 3C. FLAG CONVERSION HELPERS

def to_bool_flag(v):
    if v is None:
        return False
    s = str(v).strip().lower()
    return s in ("1", "true", "yes", "y", "current", "present")
```

```python
# 4. EXTRACT WORK EXPERIENCES

def extract_experiences(parsed):
    """
    parsed: dict from GPT_Output.
    We expect key 'Companies': list of dicts.
    We ignore education and all other sections.
    """
    if not isinstance(parsed, dict):
        return []

    companies = parsed.get("Companies") or parsed.get("companies")
    if not isinstance(companies, list):
        return []

    experiences = []
    for job in companies:
        if not isinstance(job, dict):
            continue

        company = job.get("Company Name") or job.get("company") or ""
        role    = job.get("Role") or job.get("role") or ""

        if not company and not role:
            continue

        start_raw = job.get("Start Date") or job.get("start_date") or ""
        end_raw   = job.get("End Date") or job.get("end_date") or ""

        start_date = normalize_date(start_raw)
        end_date   = normalize_date(end_raw)

        current_flag = job.get("Current_Flag") or job.get("current")
        internship_flag = job.get("Internship_Flag") or job.get("internship")

        current = to_bool_flag(current_flag)
        internship = to_bool_flag(internship_flag)


        if end_date == "Present":
            current = True

        experiences.append({
            "company": company,
            "role": role,
            "start_date": start_date,
            "end_date": end_date,
            "current": current,
            "internship": internship,
        })

    return experiences
```

```python
# Quick check on first 3 GPT outputs
for i in range(3):
    parsed = safe_load_json(df["gpt_output"].iloc[i])
    exps = extract_experiences(parsed)
    print(f"\nRow {i} experiences:")
    print(json.dumps(exps, indent=2, ensure_ascii=False))
```

```python
# 5. BUILD TRAINING RECORDS

training_data = []

for idx, row in tqdm(df.iterrows(), total=len(df)):
    resume = row["resume_text"]
    gpt_raw = row["gpt_output"]

    parsed = safe_load_json(gpt_raw)
    exps = extract_experiences(parsed)

    if not exps:

        continue

    target_obj = {"experiences": exps}

    sample = {
        "messages": [
            {
                "role": "system",
                "content": (
                    "You are an information extraction model trained for resume parsing. "
                    "Extract ONLY WORK EXPERIENCE from the resume and return strict JSON in the form:\n"
                    '{ "experiences": [ { "company": "", "role": "", "start_date": "", "end_date": "", "current": false, "internship": false } ] }\n'
                    "Rules:\n"
                    "- Do NOT include education, skills, certifications, projects, or personal details.\n"
                    "- Do NOT hallucinate missing information.\n"
                    "- Dates must be preserved or normalized, never invented.\n"
                    "- If no work experience exists, return {\"experiences\": []}."
                )
            },
            {
                "role": "user",
                "content": resume
            },
            {
                "role": "assistant",
                "content": json.dumps(target_obj, ensure_ascii=False)
            }
        ]
    }

    training_data.append(sample)

print("Valid training samples:", len(training_data))
```

```python

from unsloth import FastLanguageModel
from datasets import load_dataset
from trl import SFTTrainer, SFTConfig
from transformers import AutoTokenizer
import torch, json
```

```python
MODEL_NAME = "unsloth/Meta-Llama-3.1-8B-Instruct-bnb-4bit"
MAX_SEQ_LEN = 2048

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name      = MODEL_NAME,
    max_seq_length  = MAX_SEQ_LEN,
    load_in_4bit    = True,   # QLoRA
    load_in_8bit    = False,
    load_in_16bit   = False,
)


if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

print("Model & tokenizer loaded.")
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r                        = 32,
    lora_alpha               = 16,
    lora_dropout             = 0.05,
    target_modules           = [
        "q_proj", "k_proj", "v_proj", "o_proj",
        "gate_proj", "up_proj", "down_proj",
    ],
    use_gradient_checkpointing = "unsloth",
    random_state             = 3407,
    max_seq_length           = MAX_SEQ_LEN,
    use_rslora               = False,
    loftq_config             = None,
)

print("LoRA adapter attached.")
```

```python
def format_chat(example):

    text = tokenizer.apply_chat_template(
        example["messages"],
        tokenize = False,
        add_generation_prompt = False,
    )
    return {"text": text}

ds_formatted = ds.map(format_chat, remove_columns=[c for c in ds.column_names if c != "messages"])
print("Formatted dataset example:\n")
print(ds_formatted[0]["text"][:1000])
```

```python
BATCH_SIZE = 1
GRAD_ACCUM = 8
EPOCHS      = 3
LR          = 2e-4

config = SFTConfig(
    output_dir                    = "resume_llama_lora",
    num_train_epochs              = EPOCHS,
    per_device_train_batch_size   = BATCH_SIZE,
    gradient_accumulation_steps   = GRAD_ACCUM,
    learning_rate                 = LR,
    warmup_ratio                  = 0.03,
    logging_steps                 = 10,
    save_steps                    = 200,
    save_total_limit              = 2,
    max_seq_length                = MAX_SEQ_LEN,
    packing                       = False,
    bf16                          = False,
    fp16                          = True,
    optim                         = "adamw_8bit",
    report_to                     = "none",
)

trainer = SFTTrainer(
    model           = model,
    tokenizer       = tokenizer,
    train_dataset   = ds_formatted,
    dataset_text_field = "text",
    args            = config,
)

print("Trainer ready.")
```

```python
import textwrap

def extract_experience(resume_text: str, max_new_tokens: int = 512):
    """
    Sends a resume text to the fine-tuned model and returns parsed JSON.
    """
    system_prompt = (
        "You are an information extraction model.\n"
        "Extract ONLY work experience as strict JSON.\n"
        "Return:\n"
        "{ \"experiences\": [ {"
        "  \"company\": \"\", \"role\": \"\", "
        "  \"start_date\": \"\", \"end_date\": \"\", "
        "  \"current\": false, \"internship\": false } ] }\n"
        "If no experience: {\"experiences\": []}\n"
        "No extra text, no explanations. Dates must be EXACTLY as written."
    )

    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user",   "content": resume_text},
    ]

    prompt = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True,
    )

    inputs = tokenizer(
        prompt,
        return_tensors="pt",
        truncation=True,
        max_length=MAX_SEQ_LEN,
    ).to(base_model.device)

    with torch.no_grad():
        outputs = base_model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            temperature=0.0,
        )

    generated = tokenizer.decode(outputs[0][inputs["input_ids"].shape[1]:], skip_special_tokens=True)

    generated = generated.strip()
    try:
        start_idx = generated.find("{")
        end_idx   = generated.rfind("}")
        if start_idx != -1 and end_idx != -1:
            json_str = generated[start_idx:end_idx+1]
        else:
            json_str = generated

        return json.loads(json_str)
    except Exception as e:
        print("⚠️ JSON parse failed, returning raw text. Error:", e)
        return generated
```
